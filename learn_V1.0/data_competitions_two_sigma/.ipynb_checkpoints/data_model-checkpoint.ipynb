```python
from sklearn.metrics import log_loss
#sklearn.metrics.log_loss
import lightgbm as lgb

def evaluate_model(df, target, train_index, test_index, params):
    model = lgb.LGBMClassifier(**params)
    model.fit(df.iloc[train_index], target.iloc[train_index],verbose=10,
             eval_metric="binary_logloss",
             eval_set=[(df.iloc[train_index], target.iloc[train_index]),
             (df.iloc[test_index], target.iloc[test_index])], early_stopping_rounds=50)
    #return lgb.binary_logloss(target.iloc[test_index], model.predict_proba(df.iloc[test_index]))
    return log_loss(target.iloc[test_index], model.predict_proba(df.iloc[test_index]))
```

```python
param_grid = {
    'learning_rate': [0.05,0.03,0.01],
    'num_leaves': [50,100,150],
    'n_estimators': [ 1500,1800,2000],
    'colsample_bytree': [0.8],
    'subsample': [0.8],
    'reg_alpha': [0.5],
    'reg_lambda': [1.5,2,2.5],
    'n_jobs':[4],
    'min_child_weight':[8], 
    'min_child_samples':[8],
    'min_split_gain':[0.2],
    'objective':['binary']

}

best_eval_score = 0
for i in range(20):
    params = {k: np.random.choice(v) for k, v in param_grid.items()}
    print("params:",params)
    print("{0}/50".format(str(i)))
    score = evaluate_model(feature_train_trans_op, target_for_train, train_index, test_index, params)
    print("score:",score)
    if score < best_eval_score or best_eval_score == 0:
        best_eval_score = score
        best_params = params
print("Best evaluation logloss", best_eval_score)
```

```python
# Train model with full data
clf = lgb.LGBMClassifier(**best_params)
clf
clf.fit(feature_train_trans_op, target_for_train)
#print('Feature importances:', list(lgb.feature_importances_))
```

```python
print('Feature importances:', list(clf.feature_importances_))
```

```python
from sklearn.metrics import log_loss
#sklearn.metrics.log_loss
import lightgbm as lgb

def evaluate_model(df, target, train_index, test_index, params):
    model = lgb.LGBMClassifier(**params)
    model.fit(df.iloc[train_index], target.iloc[train_index],verbose=10,
             eval_metric="binary_logloss",
             eval_set=[(df.iloc[train_index], target.iloc[train_index]),
             (df.iloc[test_index], target.iloc[test_index])], early_stopping_rounds=50)
    #return lgb.binary_logloss(target.iloc[test_index], model.predict_proba(df.iloc[test_index]))
    return log_loss(target.iloc[test_index], model.predict_proba(df.iloc[test_index]))
```

```python
param_grid = {
    'learning_rate': [0.04,0.05,0.03],
    'num_leaves': [100],
    'n_estimators': [ 1500,1600,1700],
    'colsample_bytree': [0.8],
    'subsample': [0.8],
    'reg_alpha': [0.5],
    'reg_lambda': [2.0],
    'n_jobs':[4],
    'min_child_weight':[8], 
    'min_child_samples':[8],
    'min_split_gain':[0.2],
    'objective':['binary']

}

best_eval_score = 0
for i in range(10):
    params = {k: np.random.choice(v) for k, v in param_grid.items()}
    print("params:",params)
    print("{0}/10".format(str(i)))
    score = evaluate_model(feature_train_trans_op[feat_importance_field], target_for_train, train_index, test_index, params)
    print("score:",score)
    if score < best_eval_score or best_eval_score == 0:
        best_eval_score = score
        best_params = params
print("Best evaluation logloss", best_eval_score)
```

```python
# Train model with full data
clf = lgb.LGBMClassifier(**best_params)
clf
clf.fit(feature_train_trans_op[feat_importance_field], target_for_train)
print('Feature importances:', list(clf.feature_importances_))
```
